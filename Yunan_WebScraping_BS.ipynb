```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd 
import re
from datetime import datetime
from datetime import date
```

## Retrieve Website and Parse HTML

```python
# Identify the target website's address, i.e., URL, which is Carrefour Belgium's promotion page in this case.
Carrefour_url = 'https://drive.carrefour.be/nl/search?text=allPromotions'
# Create a response object to get the web page's HTML content
get_url = requests.get(Carrefour_url)
# Create a beautiful soup object to parse HTML text with the help of the html.parser
Carrefour_soup = BeautifulSoup(get_url.text, 'html.parser')
# Check website's response
print(get_url)
```

```python
# Get some intuition by printing out HTML
# This step is not required to build a web scraper
print(get_url.text)
print(Carrefour_soup)
# Use prettify() method to make the HTML be nicely formatted
print(Carrefour_soup.prettify())
```

## Extract, Store, and Clean Data

### Create Variables

```python
# Identify retailer's name
retailer = 'Carrefour Belgium'
# Identify the base URL 
base_url = 'https://drive.carrefour.be'
# Record the date of scraping
scraping_date = date.today()
# Create a list to store all the extracted data 
products_all = []
```

### Start by Scraping Category

```python
# Find all the categories listed on the web page
# This step is used for testing and practicing, which can be skipped for the final scraper
categories = Carrefour_soup.find('ul', {'class': 'facet__list js-facet-list Afdeling facet__list--hidden js-facet-list-hidden'}
                                ).find_all('li', {'class': 'row adding'}) 
# Loop through the categories
for category in categories: 
    
    # Get category name by extracting the text part of the <a> element
    # Strip the spaces before and after the name
    category_name = category.find('a').text.strip()
    # Get the URL, which leads to the web page of the products under the category
    category_url = category.find('a').get('href')
    # Complete category’s URL by adding the base URL
    category_url = base_url + category_url
    print(f"{category_name}'s URL is: {category_url}")
```

### Fetching all the Promotional Products by Category

```python
def fetch_all_products(soup):
    # Fetch all the products information
    # Return products_all, the list of dictionary that contains all the extracted data
    
    # Find all the categories
    categories = find_categories(soup)
    # Loop through categories
    for category in categories:
        # Fetch product by category
        # Within the fetch_products_by_category function, we scrape products page by page
        fetch_products_by_category(category)
        
    return products_all    
```

```python
def find_categories(soup):
    # Find all the categories
    
    categories = soup.find('ul', {'class': 'facet__list js-facet-list Afdeling facet__list--hidden js-facet-list-hidden'}
                          ).find_all('li', {'class': 'row adding'})
    
    return categories  
```

```python
def fetch_products_by_category(category):
    # Fetch products by category
    # Scrape all the products listed on one page
    # Go to next page if current page is not the last page
    # Break loop at the last page
    
    # Get category name by extracting the text part of <a> element
    category_name = category.find('a').text.strip()
    # Get URL, i.e., the link to the first page of products under the category
    products_page_url = base_url + category.find('a').get('href')
    # Scape products page by page only when the next page is available
    while True:
        # Retrieve the products page's HTML
        get_current_page = requests.get(products_page_url)
        # Create a beautiful soup object for the current page
        current_page_soup = BeautifulSoup(get_current_page.text, 'html.parser')
        # Run fetch_current_page_products function to get all the products listed on the current page
        fetch_current_page_products(category_name, current_page_soup)
        # Search for the next page's URL
        find_next_page = current_page_soup.find('li',{'class':'pagination-next'})
        # Get the next page's URL if the current page is not the last page
        if find_next_page.find('a').get('href') != '#':
            products_page_url = base_url + find_next_page.find('a').get('href')            
        else:
            break
    
```

```python
def fetch_current_page_products(category_name, current_page_soup):
    # Fetch all the products listed on the current page
    # Build a dictionary to store extracted data 
    # Append product information to the products_all list

    # Find all products listed on the current page
    current_page_products = current_page_soup.find_all('div', {'class':'wrapper'})
    
    # Loop through the products   
    for item in current_page_products:        
        
        # Search for the HTML elements containing product information
        prod_info = item.find('a', {'class': 'name select_item name-title select_promotion_item'}) 
        # Extract data within <a> element with get() method
        # Get unique product id  
        prod_id = prod_info.get('data-item_id')
        # Get product title, including brand, product name and quantity
        prod_title =  prod_info.get('data-item_name')
        # Get current price
        current_price = prod_info.get('data-price')
        # Get original price
        # Assign current price to original_price when original_price is not avaiable
        original_price = prod_info.get('data-base_price')
        if original_price == '':
            original_price = current_price
        # Get product link
        link = base_url + prod_info.get('href')
        # Get promotion name
        prom_name = prod_info.get('data-promotion')
        
        # Some promotional products have no promotion end date so we assign it with None as default        
        prom_end_date = None
        
        # Search for the HTML elements containing promotion information
        prom_info = item.find('div', {'class': 'tooltip-inner'})
        
        # Handle potential error using try and except       
        try:
            # Try to find promotion end date within prom_info
            # An error will be raised if prom_info is not available
            promo_end_info = prom_info.find('span', {'class': 'promo-last-text'}).text.strip()
            # The typical value of prom_end_info is 'Aanbieding geldig tot dd.mm.yyyy'
            # Extracting date out of the string
            prom_end_date = promo_end_info[-10:]
            # Convert prom_end_date from a string to a datetime object
            prom_end_date = datetime.strptime(prom_end_date, '%d.%m.%Y').date()
        except:
            # Go to the next line without crash when the try block got the error 
            pass
              
        # Assign promotion start date with None as default since the variable is not available online  
        prom_start_date = None
        
        # Create a dict to store product info
        product = {
            'retailer': retailer, # retailer variable has been created above
            'scraping_date': scraping_date, # scraping_date variable has been created above
            'product_id': prod_id, 
            'product_title': prod_title,        
            'category': category_name,            
            'original_price': original_price,
            'current_price': current_price,
            'promotion_name': prom_name,
            'promotion_start_date': prom_start_date, 
            'promotion_end_date': prom_end_date,
            'link':link
          }
        # Append product dictionary to products_all
        products_all.append(product)
       
```

```python
products_list = fetch_all_products(Carrefour_soup)
```

```python
def output(products_list):
    # Convert the list with scraped data to a data frame, drop the duplicates, and save the output as a csv file
    
    products_df = pd.DataFrame(products_list).drop_duplicates()
    print(f'There are totally {len(products_df)} products.')
    products_df.to_csv(f'Carrefour_scraper_{scraping_date}.csv', index = False) 
```

```python
output(products_list)
```
